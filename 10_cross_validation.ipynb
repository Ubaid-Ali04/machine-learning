```python
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
iris=load_iris()
model=GaussianNB()
scores=cross_val_score(model,iris.data,iris.target,cv=5,scoring='accuracy')
print('score for each fold',scores)
print('mean score',scores.mean())
print('standard deviation',scores.std())
```

```python
df=sns.load_dataset('tips')
#encode the data using for loop
for column in df.columns:
    if df[column].dtype==object:
        le=LabelEncoder()
        df[column]=le.fit_transform(df[column])
X=df[['total_bill','tip','size']]
y=df['sex']
model=RandomForestClassifier()
score=cross_val_score(model,X,y,cv=5,scoring='accuracy')
print('score for each fold',score)
print('mean score',score.mean())
print('standard deviation',score.std())
```

```python
#plot the result
plt.figure(figsize=(8,6))
plt.plot(range(1,9),score)
plt.title('cross validation')
plt.xlabel('number of folds')
plt.ylabel('accuracy')
plt.show()
```
